# Visualizing Earnings Based On College Majors

In this small project, I'll explore using the pandas plotting functionality along with the Jupyter notebook interface to explore data quickly using visualizations, commonly known as exploratory data analysis (EDA).

I'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

```
Rank - Rank by median earnings (the dataset is indexed by this column).
Major_code - Major code.
Major - Major description.
Major_category - Category of major.
Total - Total number of people with major.
Sample_size - Sample size (unweighted) of full-time.
Men - Male graduates.
Women - Female graduates.
ShareWomen - Women as share of total.
Employed - Number employed.
Median - Median salary of full-time, year-round workers.
Low_wage_jobs - Number in low-wage service jobs.
Full_time - Number employed 35 hours or more.
Part_time - Number employed less than 35 hours.
```

My objective with this project will be to answer questions like:
- What majors are the most popular?
- Do more popular majors correlate with higher pay?
- Which majors are predominantly male or female?

````python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
````

````python
with open('recent-grads.csv') as file:
    raw_data = pd.read_csv(file)

raw_data.info()
````

````python
raw_data.head()
````

````python
raw_data.describe()
````

````python
# Drop rows with NULL/NaN values
n_rows = len(raw_data.index)
recent_grads = raw_data.dropna()

print("Dropped rows: {}".format(n_rows-len(recent_grads.index)))
````

````python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample size vs Median')
````

````python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample size vs Unemployment')
````

````python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full time vs Median')
````

````python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('Fraction of Women vs Unemployment rate')
````

````python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men vs Median')
````

````python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women vs Median')
````

````python
cols = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize=(10,25))
for k in range(len(cols)):
    ax = fig.add_subplot(4,2,k+1)
    ax = recent_grads[cols[k]].hist(xrot=320)
    ax.set_title('{} disribution'.format(cols[k])), ax.set_xlabel(cols[k])
````

````python
from pandas.plotting import scatter_matrix

cols = ['Sample_size', 'Median']
ax = scatter_matrix(recent_grads[cols], figsize=(10,10))
````

````python
cols = ['Sample_size', 'Median', 'Unemployment_rate']
ax = scatter_matrix(recent_grads[cols], figsize=(15,15))
````
